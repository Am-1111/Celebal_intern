```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

dataset = sns.load_dataset('titanic')

dataset.head()
```

```python
# import seaborn as sns
# import matplotlib.pyplot as plt

sns.histplot(dataset['fare'], kde=False)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(dataset['fare'], kde=False)
```

```python
sns.jointplot(x='age', y='fare', data=dataset)
```

```python
sns.barplot(x='sex', y='age', data=dataset)
```

```python
sns.boxplot(x='sex', y='age', data=dataset, hue="survived")
```

```python
sns.violinplot(x='sex', y='age', data=dataset)
```

```python


# Create violin plot
sns.violinplot(x='sex', y='age', data=dataset)

# Add swarm plot with smaller marker size
sns.swarmplot(x='sex', y='age', data=dataset, color='black', size=3)

plt.show()
```

```python
#Expt. No. 8 Part-2
# histogram of fare
titanic_hist = dataset.fare.plot.hist(bins = 40, color = 'grey')
plt.xlabel('Fare (pounds, 1912 prices)')

plt.show(titanic_hist)
```
